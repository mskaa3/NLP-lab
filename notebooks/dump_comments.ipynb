```python
import pandas as pd
import jsonlines
from pathlib import Path
from datetime import datetime
```

```python
DATA_DIR = Path("data_nlp")
tag = 'pandoragate'
comments_path = DATA_DIR / f"raw/{tag}/comments.jsonl"
save_path = DATA_DIR / f"comments_df_{tag}.feather"
```

```python
with jsonlines.open(comments_path) as reader:
    raw_comments = [*reader]
```

```python
df = {
    "id":[],
    "topic":[],
    "publishedAt":[],
    "authorDisplayName":[],
    "textOriginal":[]
}
topic = 'pandoragate'
for i, comment in enumerate(raw_comments):
    df['id'].append(comment['snippet']['topLevelComment']['id'])
    df['topic'].append(tag)
    df['publishedAt'].append(datetime.fromisoformat(comment['snippet']['topLevelComment']['snippet']['publishedAt'].replace("Z", "")))
    df['authorDisplayName'].append(comment['snippet']['topLevelComment']['snippet']['authorDisplayName'])
    df['textOriginal'].append(comment['snippet']['topLevelComment']['snippet']['textOriginal'])

    try:
        for subcomment in comment['replies']['comments']:
            df['id'].append(subcomment['id'])
            df['topic'].append(tag)
            df['publishedAt'].append(datetime.fromisoformat(subcomment['snippet']['publishedAt'].replace("Z", "")))
            df['authorDisplayName'].append(subcomment['snippet']['authorDisplayName'])
            df['textOriginal'].append(subcomment['snippet']['textOriginal'])
    except:
        pass

df_comments = pd.DataFrame.from_dict(df)
df_comments.to_feather(save_path)
```
